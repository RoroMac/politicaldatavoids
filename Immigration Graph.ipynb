```python
def clean_address(row):
    name =  row.lower().split('/')
    if  row.lower().startswith('https') or  row.lower().startswith('http'):       
        return name[3]
    elif row.lower().startswith('www'):
        return name[1] 
    elif row.lower().startswith('facebook'):
        return name[1]
    else:
        print("Not found", row)
        return None
```

```python
#path = "Latinx2020/"
news = pd.read_csv("new.csv")
pol = pd.read_csv("political.csv")
cult = pd.read_csv("geocult.csv")
#We rename the colum Page Name to be able to concat with Group Posts
news.rename(columns={'Page Name':'name'}, inplace=True)
pol.rename(columns={'Page Name':'name'}, inplace=True)
cult.rename(columns={'Page Name':'name'}, inplace=True)
```

```python
news['com']= 'News'
pol['com'] = 'Political'
cult['com'] = 'Geo-Cultural'
print(len(news))
print(len(pol))
print(len(cult))
```

```python
df = news
df= df.append(pol, ignore_index=True)
df = df.append(cult, ignore_index=True)
```
